```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib

import matplotlib.pyplot as plt
from scipy.stats import skew
from scipy.stats.stats import pearsonr

from sklearn import linear_model


%config InlineBackend.figure_format = 'png' #set 'png' here when working on notebook
%matplotlib inline
```

# Leitura dos dados

```python
alldata = pd.read_csv("eleicoes_2006_a_2010.csv")
header_list = list(alldata.columns.values)
alldata2014 = pd.read_csv("eleicoes_2014.csv")
alldata2014.columns = header_list
alldata2014.head()

alldata = pd.concat([alldata,alldata2014])
alldata.info()
```

Avaliando as colunas, podemos perceber que algumas podem ser descartadas por não fazerem diferença na predição como ano, nome do candidato, sequencial do candidato, ocupação, cargo e estado civil.

```python
alldata = alldata.drop(['nome','sequencial_candidato', 'cargo', 'grau','ocupacao', 'estado_civil'], axis=1)
alldata.head()
```

# Correlação
Verificaremos agora a correlação das colunas restantes, a fim de identificar correlações fortes que nos permitam eliminar algumas delas.

```python
sns.heatmap(alldata.corr(), vmax=0.75, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

Avaliando o heat map, podemos perceber que as variáveis quantidade_doacoes e quantidade_doadores têm correlação forte entre si (próximo a .7), podendo eliminar uma delas. Também as variáveis total_receita, recursos_de_pessoas_juridicas, quantidade_despesas e quantidade_fornecedores possuem forte correlação com total_despesa (próximo a 0.75), podendo ficar apenas com total_despesa.

```python
alldata = alldata.drop(['quantidade_doadores', 'total_receita','recursos_de_pessoas_juridicas', 'quantidade_despesas', 'quantidade_fornecedores'], axis=1)
```

```python
alldata.head()
```

# Gerando a descrição dos dados e a soma dos valores NA

```python
print(alldata.describe())
alldata.isna().sum()
```

Como faltam muitos dados em algumas colunas, resolvemos fazer imputação. Testamos inserindo o valor da média e da mediana e a mediana mostrou ser uma melhor opção, mantendo as estatísticas descritivas mais próximas da original.

```python
alldata['media_receita'].fillna(alldata['media_receita'].median(), inplace=True)
alldata['recursos_de_outros_candidatos/comites'].fillna(alldata['recursos_de_outros_candidatos/comites'].median(), inplace=True)
alldata['recursos_de_pessoas_fisicas'].fillna(alldata['recursos_de_pessoas_fisicas'].median(), inplace=True)
alldata['recursos_proprios'].fillna(alldata['recursos_proprios'].median(), inplace=True)

#testando para ver se os dados foram inputados
alldata.isna().sum()
```

# Verificando a necessidade de normalização.

Plotando as distribuições das variáveis restantes não categóricas (forma original e já normalizada com log). Vemos que usando log, o formato  das distribuições fica mais parecido com a normal, sendo mais adequado para usar nas regularizações.

```python
df_all = pd.DataFrame(alldata)
for column in df_all:
    if df_all[column].dtype != 'object':
        matplotlib.rcParams['figure.figsize'] = (12.0, 6.0)
        alldata_hists = pd.DataFrame({"alldata:"+column:alldata[column], "alldata: log("+column+" + 1)":np.log1p(alldata[column])})
        alldata_hists.hist()
        plt.show()
```

# Normalização

```python
#separando as features da variavel alvo
alldata_feats = alldata.loc[:,'ano':'sexo']
alldata_target =  alldata[['ano','votos']]
alldata_target.head()
```

```python
#log transform skewed numeric features:
numeric_feats = alldata_feats.dtypes[alldata_feats.dtypes != "object"].index

skewed_feats = alldata[numeric_feats].apply(lambda x: skew(x)) #compute skewness
skewed_feats = skewed_feats[skewed_feats > 0.75]
skewed_feats = skewed_feats.index

alldata_feats[skewed_feats] = np.log1p(alldata_feats[skewed_feats])

#dummyfying
alldata_feats = pd.get_dummies(alldata_feats)
```

```python
alldata_feats.head()
```

# Dividindo os dados em treino (2006), validation (2010) e teste (2014)

```python
#train
train_X = alldata_feats.loc[alldata_feats['ano'] == 2006]
train_X = train_X.drop(['ano'],axis=1)
train_Y = alldata_target.loc[alldata_target['ano'] == 2006]
train_Y = train_Y.drop(['ano'],axis=1)

#validation
validation_X = alldata_feats.loc[alldata_feats['ano'] == 2010]
validation_X =validation_X.drop(['ano'],axis=1)
validation_Y = alldata_target.loc[alldata_target['ano'] == 2010]
validation_Y = validation_Y.drop(['ano'],axis=1)

#test
test_X = alldata_feats.loc[alldata_feats['ano'] == 2014]
test_X = test_X.drop(['ano'],axis=1)
test_Y = alldata_target.loc[alldata_target['ano'] == 2014]
test_Y = test_Y.drop(['ano'],axis=1)

test_Y.head()
```

# Gerando os modelos
1. Sem regularização
2. Com Ridge
3. Com lasso
4. KNN
5. SVR

```python
from sklearn.linear_model import Ridge, Lasso, LassoCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score, mean_squared_error
from math import sqrt
```

```python
def rmse_cv(model,train_X, train_y):
    rmse= np.sqrt(-cross_val_score(model, train_X, train_y, scoring="neg_mean_squared_error", cv = 5))
    return(rmse)

def plot_data(X,Y):    
    plt.plot(X,Y,'k.')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.axis([0,1,-1.5,2])

def print_coefficients(model):    
    #w = list(model.coef_)
    #w.reverse()
    #print (np.poly1d(w) + model.intercept_)
    
    # Retorna o grau do polinômio
    deg = len(model.steps[1][1].coef_)-1
    # Obtém os parâmetros estimados
    w = list(model.steps[1][1].coef_) #model.steps é usado pois o modelo é calculado usando make_pipile do scikit learn
    # Numpy tem uma função para imprimir o polinômio mas os parâmetros precisam estar na ordem inversa
    print ('Polinômio estimado para grau ' + str(deg) + ':')
    w.reverse()
    print (np.poly1d(w)+model.steps[1][1].intercept_)

def print_poly_predictions(X,Y, model):
    plot_data(X,Y)
    x_plot = np.array([i/200.0 for i in range(200)])
    X_plot = x_plot[:,np.newaxis]
    y_pred = model.predict(X_plot)
    plt.plot(x_plot,y_pred,'g-')
    plt.axis([0,1,-1.5,2])

def plot_residuals_vs_fit(X,Y, model, title):
    y_pred = model.predict(X)
    if y_pred.ndim == 1:
        y_pred=np.matrix(y_pred)
        y_pred=y_pred.transpose()
                
    res = Y - y_pred
    plt.plot(y_pred,res,'k.',color='blue',)
    plt.axhline(y=0., color='r', linestyle='-')
    plt.xlabel("predictions")
    plt.ylabel("residuals")
    plt.title(title)
    
def print_model_result(feat, target, model,title):
    predict = model.predict(feat)
    plot_residuals_vs_fit(feat, target, model,title)
    print("Results")
    print("Root Mean squared error: %.2f" % sqrt(mean_squared_error(target,predict)))
    print('R2 score: %.2f' % r2_score(target,predict))
    
def plot_rmse(series, param_name):
    series.plot(title = "Validation Error vs " + param_name)
    plt.xlabel(param_name)
    plt.ylabel("rmse")
```

# Sem regularização

```python
linear = linear_model.LinearRegression()
linear.fit(train_X, train_Y)

y_pred = linear.predict(validation_X)

print("Root Mean squared error: %.2f" % sqrt(mean_squared_error(validation_Y,y_pred)))
print('R2 score: %.2f' % r2_score(validation_Y,y_pred))


  
```

```python
Percebemos pelo RMSE e R2 score que o modelo sem regularização não é nem de longe adequado a este caso.
```

# Ridge

```python
alphas = [0.05, 0.1, 0.3, 1, 3, 5, 10, 15, 30, 50, 75]
cv_ridge = [rmse_cv(Ridge(alpha = alpha), train_X, train_Y).mean() 
            for alpha in alphas]
```

```python
cv_ridge = pd.Series(cv_ridge, index = alphas)
cv_ridge.plot(title = "Train")
plt.xlabel("alpha")
plt.ylabel("rmse")
```

```python
print('Best RMSE: ',cv_ridge.min())
best_alpha=cv_ridge.values.argmin()
print('Best alpha: ',best_alpha)
```

O melhor valor de alpha seria 8, onde temos o menor RMSE (35179.1576415288). olhando a curva, o valor 30 mantém um bom trade-off entre overfitting e flexibilidade.
Rodando o modelo com o alpha de 8 e 30 no conjunto de validação:

```python
model_ridge = Ridge(alpha=8.0)
model_ridge.fit(train_X, train_Y)
print_model_result(validation_X, validation_Y,model_ridge,"Regressão com Regularização Ridge")
```

```python
model_ridge = Ridge(alpha=30.0)
model_ridge.fit(train_X, train_Y)
print_model_result(validation_X, validation_Y,model_ridge,"Regressão com Regularização Ridge")
```

```python
Obtemos na validação RMSE de 39298 com alpha 8.0 contra 39305 votos com alpha 30. Resultado quase igual. R2 0.38 nos mostra que o modelo ainda não explica muito bem a variação na resposta. Mas, observando os gráficos, vemos que os residuais estão ao redor de zero, com algumas discrepâncias. 
```

# Lasso

```python
alphas = np.linspace(10, 100, 50,endpoint=False)
cv_lasso = [rmse_cv(Lasso(alpha = alpha), train_X,train_Y).mean() 
            for alpha in alphas ]

series_lasso = pd.Series(cv_lasso, index = alphas)
plot_rmse(series_lasso, "alpha")
```

```python
print('Best RMSE: ',series_lasso.min())
print('Best alpha: ',series_lasso.values.argmin())
```

O melhor valor de alpha seria 19, onde temos o menor RMSE (35274). Olhando a curva, o valor 50 mantém um bom trade-off entre overfitting e flexibilidade. Rodando o modelo com o alpha de 19 e 50 no conjunto de validação:

```python
model_lasso = Lasso(alpha=19.0)
model_lasso.fit(train_X, train_Y)
print_model_result(validation_X, validation_Y,model_lasso,"Regressão com Regularização Lasso")
   
```

```python
model_lasso = Lasso(alpha=50.0)
model_lasso.fit(train_X, train_Y)
print_model_result(validation_X, validation_Y,model_lasso,"Regressão com Regularização Lasso")
```

```python
Obtemos na validação RMSE de 39285.69 com alpha 19.0 contra 39258.67 votos com alpha 50. Resultado quase igual. R2 0.38 nos mostra que o modelo ainda não explica muito bem a variação na resposta. Mas, observando os gráficos, vemos que os residuais estão ao redor de zero, com algumas discrepâncias. 
```

# KNN

```python
from sklearn.neighbors import KNeighborsRegressor

neighbors = range(2,50)
cv_knn = [rmse_cv(KNeighborsRegressor(n_neighbors = k), train_X,train_Y).mean() 
            for k in neighbors ]

series_knn = pd.Series(cv_knn, index = neighbors)
plot_rmse(series_knn, "Neighbors")
```

```python
print('Best RMSE: ',series_knn.min())
print('Best number of neighbors: ',series_knn.values.argmin())
```

O melhor valor de alpha seria 21, onde temos o menor RMSE encontrado até agora(30835). Como os valores são bem parecidos a partir de 21 vizinhos, rodaremos apenas modelo com o alpha de 21 no conjunto de validação:

```python
model_knn = KNeighborsRegressor(n_neighbors = 21)
model_knn.fit(train_X, train_Y)
print_model_result(validation_X, validation_Y,model_knn,"Regressão não-paramétrica KNN")
```

Agora temos um RMSE de 36426, o menor até agora, com um R2 score de 0.47, mostrando que o modelo já explica melhor a variação na resposta. Podemos notar também, pelo gráfico, que os resíduos parecem estar menos dispersos em relação à linha 0.

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
import warnings
warnings.filterwarnings('ignore')

trees = range(5,100)
cv_randomforest = [rmse_cv(RandomForestRegressor(n_estimators = k), train_X,train_Y).mean() 
            for k in trees ]

series_randomforest = pd.Series(cv_randomforest, index = trees)
plot_rmse(series_randomforest, "Trees")
```

```python
print('Best RMSE: ',series_randomforest.min())
print('Best number of trees: ',series_randomforest.values.argmin())
```

Usando random forest, temos, pelo gráfico, que o menor erro se dá em 75 trees, já o resultado retornado pelo sklearn é 70, porém, como 7 mostra ter um valor bem alto de erro, rodaremos os dois para comparar.

```python
model_randomforest = RandomForestRegressor(n_estimators = 70)
model_randomforest.fit(train_X, train_Y)
print_model_result(validation_X, validation_Y,model_randomforest,"Random Forest")
```

```python
model_randomforest = RandomForestRegressor(n_estimators = 75)
model_randomforest.fit(train_X, train_Y)
print_model_result(validation_X, validation_Y,model_randomforest,"Random Forest")
```

Percebe-se que, realmente, o valor com 75 trees retorna melhor RMSE e R2 score, mas é o pior valor que tivemos até então entre os tipos de regressão usados.
